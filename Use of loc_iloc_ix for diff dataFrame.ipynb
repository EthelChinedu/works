```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
#bc its a dataframe method, you say 
ufo.loc[0,:]
#loc is for filtering rows and selecting columns by label. By label I mean index for row/column selection
#so for selecting row 0 and all columns, we do ufo.loc[0, :]
#we can do row 0,1,2and all columns like this
ufo.loc[[0, 1, 2], :] # I can run ufo.loc[[0:2],:] and get the same result
```

```python
#what about the column selection
ufo.loc[:, ['City', 'State']].head(10) # this will give you the column selction for 10 elements in City & State
# you can do it as a range to say give me all rows througth state in
ufo.loc[:, 'City':'State'].head(10) #we can do rows 0 thru 2, all columns = ufo.loc[0:2, 'City':'State']
```

```python
#What if we want to have city's that is oakland?
ufo.loc[ufo.City=='Oakland', 'State']
```

```python
#iloc is for filtering rows and selecting columns by integer position. The i stands for integer
#Lets assume I want column at position 0 and 3
ufo.iloc[:, [0, 3]]
```

```python
#ix allows you to mix labes and integers when doing selection. So to demonstarte this we'll pull in the drink dataset
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col='country')
```

```python
drinks.head()
```

```python
#we can select the label inside the dataset using ix like this
drinks.ix['Albania', 0]    #0 is not a label but position of the label
```

```python
#what if we want consider the position of the rows and the label, like this
drinks.ix[1, 'spirit_servings']
```
