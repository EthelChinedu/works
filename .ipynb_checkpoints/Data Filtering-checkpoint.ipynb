```python
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
movies[movies.duration >= 200] # filter criteria to o/p movies with 200 min duration
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')] #what happens here is that python would look for the movies above 200 duration and only the genre. 
                                                                #the parentheis help python to carryout an evaluation order 
```

```python
#if we do and OR in this form |, we would have loads of movies bc some are drama but doesnt meet the condition wich is above 200 duration
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]
```
